Para ver la lista de argumentos que acepta la función Interface podemos mirar la documentación: https://www.gradio.app/docs

```python
import numpy as np
import gradio as gr


def reverse_audio(audio):
    sr, data = audio
    reversed_audio = (sr, np.flipud(data))
    return reversed_audio


mic = gr.Audio(sources="microphone", type="numpy", label="Speak here...")
gr.Interface(reverse_audio, mic, "audio").launch()
```

```python
import numpy as np
import gradio as gr

notes = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]


def generate_tone(note, octave, duration):
    sr = 48000
    a4_freq, tones_from_a4 = 440, 12 * (octave - 4) + (note - 9)
    frequency = a4_freq * 2 ** (tones_from_a4 / 12)
    duration = int(duration)
    audio = np.linspace(0, duration, duration * sr)
    audio = (20000 * np.sin(audio * (2 * np.pi * frequency))).astype(np.int16)
    return (sr, audio)


gr.Interface(
    generate_tone,
    [
        gr.Dropdown(notes, type="index"),
        gr.Slider(minimum=4, maximum=6, step=1),
        gr.Textbox(type="text", value=1, label="Duration in seconds"),
    ],
    "audio",
).launch()
```

Vamos a construir una interface que permita hacer una demo de speech-recognition

```python
from transformers import pipeline
import gradio as gr

model = pipeline("automatic-speech-recognition")


def transcribe_audio(mic=None, file=None):
    if mic is not None:
        audio = mic
    elif file is not None:
        audio = file
    else:
        return "You must either provide a mic recording or a file"
    transcription = model(audio)["text"]
    return transcription


gr.Interface(
    fn=transcribe_audio,
    inputs=[
        gr.Audio(sources="microphone", type="filepath"),
        gr.Audio(sources="upload", type="filepath"),
    ],
    outputs="text",
).launch()
```

```python
from pathlib import Path
import torch
import gradio as gr
from torch import nn

LABELS = Path("/content/drive/MyDrive/Colab Notebooks/Proyectos/class_names.txt").read_text().splitlines()

model = nn.Sequential(
    nn.Conv2d(1, 32, 3, padding="same"),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(32, 64, 3, padding="same"),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(64, 128, 3, padding="same"),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Flatten(),
    nn.Linear(1152, 256),
    nn.ReLU(),
    nn.Linear(256, len(LABELS)),
)
state_dict = torch.load("/content/drive/MyDrive/Colab Notebooks/Proyectos/pytorch_model.bin")
model.load_state_dict(state_dict, strict=False)
```

```python
model.eval()
```

```python
import numpy as np
```

```python
def predict(im):
    image_array = np.array(im["composite"], dtype=np.float32)
    # Ensure the image is 2D (grayscale)
    if image_array.ndim == 3:
        image_array = image_array[..., 0]

    from PIL import Image
    image = Image.fromarray(image_array).resize((28, 28))
    image_array = np.array(image, dtype=np.float32)

    # Normalize the pixel values to the range [0, 1]
    image_array = image_array / 255.0

    # Convert the numpy array to a PyTorch tensor
    x = torch.tensor(image_array).unsqueeze(0).unsqueeze(0)
    with torch.no_grad():
        out = model(x)
    probabilities = torch.nn.functional.softmax(out[0], dim=0)
    values, indices = torch.topk(probabilities, 5)
    return {LABELS[i]: v.item() for i, v in zip(indices, values)}
```

```python
interface = gr.Interface(
    predict,
    inputs=gr.Sketchpad(),
    outputs="label",
    theme="huggingface",
    title="Sketch Recognition",
    description="Who wants to play Pictionary? Draw a common object like a shovel or a laptop, and the algorithm will guess in real time!",
    article="<p style='text-align: center'>Sketch Recognition | Demo Model</p>",
    live=True,
)
interface.launch(debug=True)
```

### More custimized

```python
import requests
import tensorflow as tf
from PIL import Image

import gradio as gr

inception_net = tf.keras.applications.MobileNetV2()  # load the model

# Download human-readable labels for ImageNet.
response = requests.get("https://git.io/JJkYN")
labels = response.text.split("\n")


def classify_image(inp):
    image_array = np.array(inp, dtype=np.uint8)

    # Asegurarse de que la imagen sea 3D (contiene canales de color)
    if image_array.ndim == 2:  # Si la imagen es en escala de grises, conviértela a RGB
        image_array = np.stack([image_array] * 3, axis=-1)

    # Redimensionar la imagen al tamaño esperado por MobileNetV2
    image = Image.fromarray(image_array).resize((224, 224))
    image_array = np.array(image, dtype=np.float32)

    # Normalizar los valores de los píxeles según lo esperado por MobileNetV2
    image_array = tf.keras.applications.mobilenet_v2.preprocess_input(image_array)

    # Agregar una dimensión extra para representar el lote (batch)
    image_array = np.expand_dims(image_array, axis=0)  # Shape: [1, 224, 224, 3]

    prediction = inception_net.predict(image_array).flatten()
    return {labels[i]: float(prediction[i]) for i in range(1000)}


image = gr.Image(height=224, width=224)
label = gr.Label(num_top_classes=3)

title = "Gradio Image Classifiction + Interpretation Example"
gr.Interface(
    fn=classify_image, inputs=image, outputs=label, title=title
).launch(debug=True)
```
