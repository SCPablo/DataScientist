### Ejercicio 1


Escribe una clase llamada Libro que represente un libro en una biblioteca. La clase debe tener los siguientes atributos y métodos:

*Atributos:*

* titulo (cadena de texto)
* autor (cadena de texto)
* anio (entero)
* disponible (booleano, por defecto True)

*Métodos:*

* prestar que marque el libro como no disponible (disponible = False)
* devolver que marque el libro como disponible (disponible = True)
* mostrar_info que imprima la información del libro (título, autor, año, disponibilidad)

```python
#### Rellenar
```

```python
class Libro:
    def __init__(self, titulo, autor, anio):
        self.titulo = titulo
        self.autor = autor
        self.anio = anio
        self.disponible = True

    def prestar(self):
        if self.disponible:
            self.disponible = False
        else:
            print(f"El libro '{self.titulo}' ya está prestado.")

    def devolver(self):
        if not self.disponible:
            self.disponible = True
        else:
            print(f"El libro '{self.titulo}' ya está disponible.")

    def mostrar_info(self):
        disponible_str = "Sí" if self.disponible else "No"
        print(f"Título: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Año: {self.anio}")
        print(f"Disponible: {disponible_str}")

# Ejemplo de uso:
libro1 = Libro("1984", "George Orwell", 1949)
libro1.mostrar_info()
# Salida:
# Título: 1984
# Autor: George Orwell
# Año: 1949
# Disponible: Sí

libro1.prestar()
libro1.mostrar_info()
# Salida:
# Título: 1984
# Autor: George Orwell
# Año: 1949
# Disponible: No

libro1.devolver()
libro1.mostrar_info()
# Salida:
# Título: 1984
# Autor: George Orwell
# Año: 1949
# Disponible: Sí
```

### Ejercicio 2 (Herencia)


Crea una clase Vehiculo con un método arrancar. Luego crea una clase Coche que herede de Vehiculo y añada un método tocar_bocina.

```python
class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

    def arrancar(self):
        return f"El {self.marca} {self.modelo} está arrancando."

class Coche(Vehiculo):
    def tocar_bocina(self):
        return f"El {self.marca} {self.modelo} está tocando la bocina."

# Crear una instancia de la clase Coche
mi_coche = Coche("Toyota", "Corolla")
print(mi_coche.arrancar())
print(mi_coche.tocar_bocina())
```

### Ejercicio 3 (Sobre escribir métodos)

Crea una clase Empleado con un método calcular_salario. Luego crea una clase Gerente que herede de Empleado y sobrescriba el método calcular_salario para añadir un bono.

```python
class Empleado:
    def __init__(self, nombre, salario_base):
        self.nombre = nombre
        self.salario_base = salario_base

    def calcular_salario(self):
        return self.salario_base

class Gerente(Empleado):
    def __init__(self, nombre, salario_base, bono):
        super().__init__(nombre, salario_base)
        self.bono = bono

    def calcular_salario(self):
        return self.salario_base + self.bono

# Crear una instancia de la clase Empleado
empleado = Empleado("Juan", 3000)
print(f"Salario de {empleado.nombre}: {empleado.calcular_salario()}")

# Crear una instancia de la clase Gerente
gerente = Gerente("Ana", 5000, 1000)
print(f"Salario de {gerente.nombre}: {gerente.calcular_salario()}")
```

### Ejercicio 4 (Polimorfismo)

Crea una interfaz Instrumento con un método tocar. Luego crea clases Guitarra y Piano que implementen esta interfaz y definan el método tocar.

```python
from abc import ABC, abstractmethod

class Instrumento(ABC):
    @abstractmethod
    def tocar(self):
        pass

class Guitarra(Instrumento):
    def tocar(self):
        return "La guitarra está sonando."

class Piano(Instrumento):
    def tocar(self):
        return "El piano está sonando."

# Crear instancias de las clases Guitarra y Piano
mi_guitarra = Guitarra()
mi_piano = Piano()

# Usar polimorfismo
def tocar_instrumento(instrumento):
    print(instrumento.tocar())

tocar_instrumento(mi_guitarra)
tocar_instrumento(mi_piano)
```
